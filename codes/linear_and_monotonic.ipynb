```python
import numpy as np
from scipy.stats import pearsonr,spearmanr,kendalltau
from matplotlib import pyplot as plt
import pandas as pd

x = np.linspace(0,20)

fig, ax = plt.subplots(1,2,figsize=(15,5))

##Linear
y1 = 5*x
y2 = -5*x

#calculating coeffs 
cor1, _ = pearsonr(x, y1)
rho1, _ = spearmanr(x, y1)
tau1, _ = kendalltau(x, y1)

cor2, _ = pearsonr(x, y2)
rho2, _ = spearmanr(x, y2)
tau2, _ = kendalltau(x, y2)

stats1 = (f'Pearson correlation: {cor1:.2f}\n'
          f'Spearman correlation: {rho1:.2f}\n'
          f'Kendal correlation: {tau1:.2f}')
stats2 = (f'Pearson correlation: {cor2:.2f}\n'
          f'Spearman correlation: {rho2:.2f}\n'
          f'Kendal correlation: {tau2:.2f}')

bbox = dict(boxstyle='square', fc='white', ec='black', alpha=0.5)

#plotting
ax[0].text(0.05, 0.8, stats1, fontsize=13, bbox=bbox,
            transform=ax[0].transAxes, horizontalalignment='left')
ax[0].scatter(x,y1,color = 'g')
ax[0].grid()
ax[0].set_xlabel("x", fontsize = 14)
ax[0].set_ylabel("y",fontsize = 14)

ax[1].text(0.95,0.8, stats2, fontsize=13, bbox=bbox,
            transform=ax[1].transAxes, horizontalalignment='right')
ax[1].scatter(x,y2,color = 'g')
ax[1].grid()
ax[1].set_xlabel("x",fontsize = 14)
ax[1].set_ylabel("y",fontsize = 14)

plt.show()
```
